# Part I. Root-finding. Newton's iteration.

Write a function which performs Newton's iteration for a given function $f(x)$ with known derivative $f'(x)$. Your function should find the root of $f(x)$ with a predefined absolute accuracy $\epsilon$. 

```python
def newton_iteration(f, fder, x0, eps=1e-5, maxiter=1000):
    """Find a root of $f(x) = 0$ via Newton's iteration starting from x0.
    
    Parameters
    ----------
    f : callable
        The function to find a root of.
    fder : callable
        The derivative of `f`.
    x0 : float
        Initial value for the Newton's iteration.
    eps : float
        The target accuracy. 
        The iteration stops when the distance between successive iterates is below `eps`.
        Default is 1e-5.
    maxiter : int
        The maximum number of iterations (default is 1000.)
        Iterations terminate if the number of iterations exceeds `maxiter`.
        This parameter is only needed to avoid infinite loops if iterations wander off.
        
    Returns
    -------
    x : float
        The estimate for the root.
    niter : int
        The number of iterations.
    """
    # ... ENTER YOUR CODE HERE ...

    nx = x0 - f(x0) / fder(x0)
    cont = 0
    
    while (abs(nx - x0) > eps): 
        x0 = nx
        nx = x0 - f(x0) / fder(x0)
        cont += 1 
        
        if cont > maxiter: 
            break
    
    return nx, cont
```

```python
def newton_iteration_fixed_poinit(f, fder, x0, alpha, eps=1e-5, maxiter=1000):
    
    nx = x0 - alpha * f(x0)
    cont = 0
    
    while (abs(nx - x0) > eps): 
        x0 = nx
        nx = x0 - alpha * f(x0)
        cont += 1 
        
        if cont > maxiter: 
            nx = 'Nan'
            break
    
    return nx, cont
```

# Part III. Newton's fractal.

(Not graded). 

Consider the equation

$$
x^3 = 1
$$

It has three solutions in the complex plane, $x_k = \exp(i\, 2\pi k/ 3)$, $k = 0, 1, 2$.

The Newton's iterations converge to one of these solutions, depending on the starting point in the complex plane (to converge to a complex-valued solution, the iteration needs a complex-valued starting point).

Plot the \emph{basins of attraction} of these roots on the complex plane of $x$ (i.e., on the plane $\mathrm{Re}x$ -- $\mathrm{Im}x$). To this end, make a series of calculations, varying the initial conditions on a grid of points. 
Color the grid in three colors, according to the root, to which iterations converged.

```python
import cmath

def newton_frac(x0):
    # Perform Newton's iteration for a specific fractional function
    
    eps = 1e-5  # Desired accuracy
    maxiter = 1000  # Maximum number of iterations
    
    nx = x0 - (x0**3 - 1) / (3 * x0**2)  # Compute the next value of x
    cont = x0  # Store the initial value of x
    
    for _ in range(maxiter):
        if abs(nx - cont) <= eps:  # Check if the desired accuracy is reached
            break
        
        if abs(nx - cont) > 1e2:  # Check if the iteration diverges
            return 0
        
        cont = nx  # Update the current value of x
        nx = nx - (nx**3 - 1) / (3 * nx**2)  # Compute the next value of x using the fractional function
        
    if abs(nx - cmath.exp(complex(0, 2 * np.pi * 1/3))) < abs(nx - cmath.exp(complex(0, 2 * np.pi * 2/3))):
        return 1  # Return 1 if the final value of x is closer to exp(2πi/3)
    else:
        return 2  # Return 2 if the final value of x is closer to exp(4πi/3)
```

```python
x = np.linspace(-1.5, 1.5, 200)  # Generate an array of x values within the specified range
y = np.linspace(-1.5, 1.5, 200)  # Generate an array of y values within the specified range

for x0 in x:
    for y0 in y:
        fr = newton_frac(complex(x0, y0))  # Compute the fractional iteration for the complex number (x0, y0)
        if fr == 1:
            plt.plot(x0, y0, marker='o', markersize=3, color='#00FF00')  # Plot green markers for fr=1
        elif fr == 2:
            plt.plot(x0, y0, marker='o', markersize=3, color='#FF0000')  # Plot red markers for fr=2
        elif fr == 3:
            plt.plot(x0, y0, marker='o', markersize=3, color='#000000')  # Plot red markers for fr=2
        
roots_x = []
roots_y = []
for i in range(3):
    z = cmath.exp(complex(0, 2 * np.pi * i / 3))  # Compute the roots of unity
    roots_x.append(z.real)
    roots_y.append(z.imag)

plt.scatter(roots_x, roots_y, c='#00FF00')  # Scatter plot the roots with green color
plt.show()
```
