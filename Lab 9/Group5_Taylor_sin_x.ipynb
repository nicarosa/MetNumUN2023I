# **Taylor series of $sin(x)$**

$$ sin(x) = \sum_{n=0}^{\infty} (-1)^{n}\frac{x^{2n+1}}{(2n+1)!} $$

# **Interactive Taylor Series of** $$f(x) = sin(x)$$

#What is value of the partial Taylor series of $sin(x)$ up to the  $10$-th term of $sin(x)$ about $0$ (Maclaurin  series) for $x=2$

$$ sin(x) \approx \sum_{i=0}^{n} (-1)^{i}\frac{x^{2i+1}}{(2i+1)!}  = x - \frac{x^3}{3!}+ \frac{x^5}{5!} +\cdots+ (-1)^{n}\frac{x^{2n+1}}{(2n+1)!}$$

```python
import math

n = 10
x = 2
sin_of_x = 0

for i in range(n+1):
    sin_of_x += ((-1)**i)*x**(2*i+1)/math.factorial(2*i+1) 
    print('sin_of_x ≈ ∑_{i=1}^'+str(i),'[(-1)^n][(x^(2i+1))/((2i+1)!)] =',sin_of_x, )  
print('math.sin(x) = ',math.sin(x))
```

# What is the minimun $n$ for which the error between the Taylor series for $sin(x)$ about $0$ (Maclaurin  series) for $x$ and *math.exp(x)*, is in absolute value less than  $\epsilon$.

$$\min_{n \in \mathbb{N}}\left|\sum_{i=0}^{n} \frac{x^i}{i!} - math.exp(x) \right| < \epsilon$$

```python
import math

x = -4
epsilon = 10**(-5)

sin_of_x = 0

print("n".rjust(10)," ","∑_{i=1}^n  [(-1)^n][(x^(2i+1))/((2i+1)!)]".center(21)," ","delta".center(21)," ","error".center(14)," ","epsilon".center(20))
error = 1
i = 0

while  epsilon <= error:
  val = ((-1)**i)*x**(2*i+1)/math.factorial(2*i+1) 
  sin_of_x += val
  error = abs(math.sin(x) - sin_of_x)
  print(format(i, '10'),"  ", format(sin_of_x, '.18f'),"  ",format(val, '.17f')," ",format(error, '.17f'),"  ","{:.0e}".format(epsilon)) 
  i = i+1


print("n =",i-1) 
print("x =",x) 
print('sin_of_x ≈ ∑_{i=0}^'+str(i-1),'x^i/i! =',sin_of_x)  
print("epsilon =","{:.0e}".format(epsilon)) 
print('math.exp(x) = ',math.exp(x))
```

**Solution** for $sin(-x)$  calculate $sin(-x)$ and return $1-sin(x)$

$$sin(-x)  = -sin(x) $$

```python
# Returns Min n, |x^{n}/n!| < epsilon and ∑_{i=0}^{n} x^i/i!

import math

x = - 43
epsilon = 10**(-5)

xo = x

if x < 0 :
   x = - x

delta = 1
sin_of_x = 0
i = 0

print("n".rjust(10)," ","∑_{i=0}^n x^i/i!".center(20)," ","|delta|".center(20)," ","epsilon".center(9))
while  epsilon <= abs(delta):
  delta = ((-1)**i)*x**(2*i+1)/math.factorial(2*i+1) 
  sin_of_x += delta
  print(format(i, '10'),"  ", format(sin_of_x, '.17f'),"  ",format( abs(delta), '.17f'),"  ","{:.0e}".format(epsilon)) 
  i = i+1

if x != xo: 
  sin_of_x = 1/sin_of_x
  x = xo
  
print("n = ",i-1) 
print('x = ',x)
print('sin_of_x ≈',sin_of_x)
print("epsilon =","{:.0e}".format(epsilon))     
print('math.sin(x) = ',math.sin(x))
```

```python
import time

start = time.time()

x = 700
epsilon = 10**(-20)

xo = x

if x < 0 :
   x = - x

delta = 1
sin_of_x = 1
i = 1

while  epsilon <= abs(delta):
  delta *= x/i 
  sin_of_x += delta
  i = i+1

if x != xo: 
  sin_of_x = 1 / sin_of_x
  x = xo

end = time.time()

print("n = ",i-1) 
print('x = ',x)
print('sin_of_x ≈',sin_of_x)
print("epsilon =","{:.0e}".format(epsilon))     
print('math.sin(x) = ',math.sin(x))
print('time = ',end - start)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sympy import Symbol, sin

def taylor_approximation(func, center, degree, values):
    taylor_series = [func.subs(x, center)]
    for n in range(1, degree + 1):
        derivative = func.diff(x, n)
        term = derivative.subs(x, center) / math.factorial(n)
        taylor_series.append(term * (values - center) ** n)
    return sum(taylor_series)

x = Symbol('x')
sin_func = sin(x)
values = np.arange(-5, 5, 0.1)
p_sin = np.sin(values)
t_sin1 = [taylor_approximation(sin_func, 0, 1, v) for v in values]
legends = ['sin(x)', 'Taylor 1 (linear)']

fig, ax = plt.subplots()
ax.plot(values, p_sin, color='red')
ax.plot(values, t_sin1, color='blue')

ax.set_ylim([-5, 5])
ax.axhline(y=0.0, xmin=-5.0, xmax=5.0, color='black')
ax.axvline(x=0.0, ymin=-10.0, ymax=10.0, color='black')
ax.legend(legends)

plt.show()
```
